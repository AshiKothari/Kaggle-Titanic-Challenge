```python
import pandas as pd
import numpy as np
data=pd.read_csv('train.csv')
testdata=pd.read_csv('test.csv')
data
```

```python
data.info()
```

```python
column_drop=["PassengerId","Name","Ticket","Cabin","Embarked"]
dt=data.drop(column_drop,axis=1)
dtest=testdata.drop(column_drop,axis=1)
dt.head(n=5)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dt["Sex"]=le.fit_transform(dt["Sex"])
dtest["Sex"]=le.fit_transform(dtest["Sex"])
dt.head()
```

```python
dt.info()
dtest.info()
```

```python
dt=dt.fillna(dt["Age"].mean())
dtest=dtest.fillna(dtest["Age"].mean())
ic = ['Pclass',"Sex","Age","SibSp","Parch","Fare"]
oc = ["Survived"]
x = dt[ic]
y = dt[oc]
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(criterion='entropy',max_depth=5)
tree.fit(x,y)
```

```python
s=tree.predict(dtest[ic])
s
```

```python
print(type(s))
```

```python
testoutput = pd.DataFrame({"PassengerID":testdata['PassengerId'],"Survived":s})
testoutput.to_csv('output.csv', sep=',', index=False)
testoutput
```
